# Sine 함수

## 변형 1
원래 sound에서 다룬 기본 type에서 freq값을 100 → 200 로 늘리기
- 태극문양 반복이 더 빽빽해짐. 같은 시간 내에 표현되는 태극문양이 두배로

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.axes_grid1 import make_axes_locatable
import IPython.display as ipd
import numpy as np
%matplotlib notebook
from scipy.signal import lfilter
```

```python
# parameter setting
amp = 1          # range [0.0, 1.0]
sr = 10000       # sampling rate, Hz
dur = 0.5        # in seconds
freq = 200.0     # sine frequency, Hz
```

```python
# generate time
t = np.arange(1, sr * dur+1)/sr
```

```python
# generate phase
theta = t * 2*np.pi * freq
```

```python
# generate signal by cosine-phasor
s = np.sin(theta)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t[0:1000], s[0:1000], '.')
ax.set_xlabel('time (s)')
ax.set_ylabel('real')
ipd.Audio(s, rate=sr)
```

## 변형 2
위 type 에서 t와 s 개수를 줄여보기
- 더 듬성한 형태 / 표현되는 시간도 범위 줄어듬

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t[0:200], s[0:200], '.')
ax.set_xlabel('time (s)')
ax.set_ylabel('real')
ipd.Audio(s, rate=sr)
```

# Exponential 함수

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.axes_grid1 import make_axes_locatable
import IPython.display as ipd
import numpy as np
%matplotlib notebook
from scipy.signal import lfilter
```

```python
# parameter setting
amp = 1          # range [0.0, 1.0]
sr = 10000       # sampling rate, Hz
dur = 0.5        # in seconds
freq = 100.0     # sine frequency, Hz
```

```python
# generate time
t = np.arange(1, sr * dur+1)/sr
```

```python
# generate phase
theta = t * 2*np.pi * freq
```

```python
# generate signal by complex-phasor
c = np.exp(theta*1j)
```

```python
c.real
```

```python
c.imag
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(t[0:1000], c.real[0:1000], c.imag[0:1000], '.')
ax.set_xlabel('time (s)')
ax.set_ylabel('real')
ax.set_zlabel('imag')
```
